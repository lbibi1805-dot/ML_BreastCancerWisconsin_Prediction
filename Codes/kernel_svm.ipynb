# KNN_Neighbours

## Importing the libraries

```python
import pandas as pd
import numpy as np
from pprint import pprint
```

## Importing the dataset

```python
dataset = pd.read_csv('../Dataset/breast_cancer_wisconsin.csv')
```

```python
print("Dataset Information");
print("__________________________")
pprint(dataset.info());
print("\n")
print("Number of null data:\n")
print("__________________________")
pprint(pd.isnull(dataset).sum())
```

Proccess null value:

```python
threshold = 0.05;

if dataset.isna().sum().sum() / dataset.size < threshold:
    dataset = dataset.dropna()
else:
    for col in dataset.columns:
        if dataset[col].dtype in ['float64', 'int64']:
            # Điền NaN bằng mean cho dữ liệu số
            dataset[col] = dataset[col].fillna(dataset[col].mean())
        else:
            # Điền NaN bằng giá trị phổ biến nhất cho dữ liệu dạng object/categorical
            dataset[col] = dataset[col].fillna(dataset[col].mode()[0])

print("Number of null data:\n")
print("__________________________")
pprint(pd.isnull(dataset).sum())
```

## Declare features and dependant variables

On the features, remove the "Sample code number" because it is not relevant to the prediction

```python
X = dataset.iloc[:,1:-1].values;
pprint(X)
```

```python
y = dataset.iloc[:, -1].values;
pprint(y);
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split;

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0);

print("DataSet Splitting:\n");
print("_______________________________")
print("X_train: ", np.size(X_train));
print("X_test: ", np.size(X_test));
print("y_train:", np.size(y_train));
print("y_test", np.size(y_test));
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

print("Feature Scaling Applied Successfully!")
print("_____________________________________")
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("\nTraining set - First 5 samples after scaling:")
pprint(X_train[:5])
print("\nTest set - First 5 samples after scaling:")
pprint(X_test[:5])
```

## Training the Kernel SVM model on the Training set

```python
from sklearn.svm import SVC
classifier = SVC(kernel='rbf', random_state=0);
classifier.fit(X_train, y_train)
```

## Predicting the Test set results

```python
y_pred = classifier.predict(X_test)
pprint(y_pred)
```

## Making the Confusion Matrix

```python
from sklearn.metrics import confusion_matrix;

cm = confusion_matrix(y_test, y_pred);
pprint(cm)
```

## Visualizing the Training set results

```python
from matplotlib.colors import ListedColormap
import matplotlib.pyplot as plt
from sklearn.svm import SVC

# We'll use only the first two features for visualization (since we can't visualize 9D)
X_set, y_set = X_train[:, [0, 1]], y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))

# Create a temporary SVM classifier with only 2 features using same parameters as main classifier
classifier_2d = SVC(kernel='rbf', random_state=0)
classifier_2d.fit(X_set, y_set)

plt.figure(figsize=(10, 8))
plt.contourf(X1, X2, classifier_2d.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(['#FA8072', '#1E90FF']))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], 
                c = ListedColormap(['#FA8072', '#1E90FF'])(i), label = f'Class {j}')
plt.title('Kernel SVM with RBF (Training set)')
plt.xlabel('Feature 1 (Clump Thickness)')
plt.ylabel('Feature 2 (Uniformity of Cell Size)')
plt.legend()
plt.show()

print("Visualization uses SVM with RBF kernel (same as main classifier)")
```

## Testing different random_state values to find optimal SVM performance

```python
random_state_test = 1
if random_state_test == 1:
    # Test different random_state values to find optimal SVM performance
    random_states = [0, 1, 42, 100, 123, 999]
    X_set, y_set = X_train[:, [0, 1]], y_train
    
    # Store results for comparison
    results = []
    
    fig, axes = plt.subplots(2, 3, figsize=(15, 10))
    axes = axes.ravel()

    for idx, rs in enumerate(random_states):
        # Create SVM classifier with different random_state
        classifier_rs = SVC(kernel='rbf', random_state=rs)
        classifier_rs.fit(X_set, y_set)
        
        # Calculate accuracy on 2D data
        accuracy_2d = classifier_rs.score(X_set, y_set)
        results.append((rs, accuracy_2d))
        
        # Create mesh for visualization
        X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.1),
                            np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.1))
        
        # Plot decision boundary
        axes[idx].contourf(X1, X2, classifier_rs.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
                          alpha = 0.75, cmap = ListedColormap(['#FA8072', '#1E90FF']))
        
        # Plot data points
        for i, j in enumerate(np.unique(y_set)):
            axes[idx].scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], 
                            c = ListedColormap(['#FA8072', '#1E90FF'])(i), label = f'Class {j}')
        
        axes[idx].set_title(f'SVM (random_state={rs})\nAccuracy: {accuracy_2d:.3f}')
        axes[idx].set_xlabel('Feature 1')
        axes[idx].set_ylabel('Feature 2')
        axes[idx].legend()

    plt.tight_layout()
    plt.show()
    
    # Print comparison results
    print("Random State Comparison Results (on 2D data):")
    print("=" * 45)
    for rs, acc in results:
        print(f"Random State {rs:3d}: Accuracy = {acc:.4f}")
    
    # Find best random state
    best_rs, best_acc = max(results, key=lambda x: x[1])
    print(f"\nBest Random State: {best_rs} with accuracy {best_acc:.4f}")
    
    # Now test on full 9D data for more comprehensive comparison
    print("\n" + "=" * 50)
    print("Full 9D Data Comparison:")
    print("=" * 50)
    
    full_results = []
    for rs in random_states:
        classifier_full = SVC(kernel='rbf', random_state=rs)
        classifier_full.fit(X_train, y_train)
        
        # Test on both training and test sets
        train_acc = classifier_full.score(X_train, y_train)
        test_acc = classifier_full.score(X_test, y_test)
        full_results.append((rs, train_acc, test_acc))
        
        print(f"Random State {rs:3d}: Train Acc = {train_acc:.4f}, Test Acc = {test_acc:.4f}")
    
    # Find best random state based on test accuracy
    best_rs_full, best_train, best_test = max(full_results, key=lambda x: x[2])
    print(f"\nBest Random State for 9D: {best_rs_full} (Test Acc: {best_test:.4f})")
    
else:
    print("Random state comparison skipped. Set random_state_test = 1 to run comparison.")
```

```python
# Visualize Test set results using the same 2 features for SVM
X_set, y_set = X_test[:, [0, 1]], y_test
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))

plt.figure(figsize=(10, 8))
plt.contourf(X1, X2, classifier_2d.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(['#FA8072', '#1E90FF']))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], 
                c = ListedColormap(['#FA8072', '#1E90FF'])(i), label = f'Class {j}')
plt.title('Kernel SVM with RBF (Test set)')
plt.xlabel('Feature 1 (Clump Thickness)')
plt.ylabel('Feature 2 (Uniformity of Cell Size)')
plt.legend()
plt.show()

print("Test set visualization uses SVM with RBF kernel (same as main classifier)")
```

## Computing the accuracy with k-Fold Cross Validation

```python
from sklearn.model_selection import cross_val_score

accuracies = cross_val_score(estimator=classifier, X=X_train, y=y_train, cv=10)
print("Accuracies:  ", format(accuracies.mean() * 100), " %")
print("Standard Deviation:  ", format(accuracies.std() * 100), " %")
```
